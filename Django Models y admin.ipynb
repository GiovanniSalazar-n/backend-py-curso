## cear el db con 500 productos 

dentro de python manage.py shell corremos la siguiente secuencia

```python
products_data=[]
for i in range(111,501):
    new_data={"title":"Product {}".format(i) , "price":i*100+99.99}
    products_data.append(new_data)
```

```python
from ecomerce.models import ProductModel
new_objects = []
for product_data in products_data:
    print(product_data)
    new_objects.append(ProductModel(**product_data))
```

```python
ProductModel.objects.bulk_create(new_objects,ignore_conflicts=True)
```

esto ya agrego todos los 500 productos a el db 

salimos de el shell creamos el fixture  corriendo el siguiente comando en la carpeta del repositorio 

 python manage.py dumpdata ecomerce --indent 4 --format json > ecomerce/fixtures/products/500Products.json
