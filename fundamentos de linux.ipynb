## estructura de proyectos en django

- manage.py : archivo que permite interactuar con django
- ocnfig__init__py: archivo vacio que deeberia ser considerado como un paquete de python 
- config/settings.py: configuracion o ajustes para el proyecto en django 
- config/urls.py: la declaracion de los urls para el proyecto django . es como una tabla de contenido de edpoints del sitio web
- config/asgi.py: un punto de entrada para servidores web compatibles con asgi para despelegar proyecto
- config/wsgi.py: un punto de entrada para servidores web compatibles con wsgi para despelegar proyecto

### archivos de aplicacion 


- admin.py:modelos relacionados con el django admin 
- app.py :contiene el codigo de la configuracion de la app 
- models.py: define tus modelos y son las tablas en la base de datos para almacenar la informacion y - define relaciones 
- test.py :escribe tes para la app 
- views.py: define la informacion que sera mandada al template del front end


## haz tus cambios visibles en el proyecto 


Lo anteriorn fueron mis notas ahora agregtare las modificaciones al proyecto de django 

```python
#views.py modificado 
import os

from django import get_version
from django.conf import settings
from django.shortcuts import render

from .clases import Alumno, Curso  


def home(request):
    alumno = Alumno("Giovanni", "Ingenier√≠a")
    curso = Curso("Fundamentos de Linux e Introducci√≥n a Django", "90 minutos")

    context = {
        "debug": settings.DEBUG,
        "django_ver": get_version() + " Probando",
        "python_ver": os.environ["PYTHON_VERSION"] + " m√°s cambios",

      
        "alumno_nombre": alumno.nombre,
        "alumno_carrera": alumno.carrera, 
        "curso_nombre": curso.nombre,
        "curso_duracion": curso.duracion,
    }

    return render(request, "pages/home.html", context)
```

```python
#clases.py
class Alumno:
    def __init__(self, nombre, carrera):
        self.nombre = nombre
        self.carrera = carrera


class Curso:
    def __init__(self, nombre, duracion):
        self.nombre = nombre
        self.duracion = duracion
```
